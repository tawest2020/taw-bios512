## Question 1

```r
table_a <- tibble(
  SKU = c(102345, 104567, 108912, 109876, 112233),
  Fruit = c("Apple", "Orange", "Mango", "Blueberry", "Watermelon"),
  Color = c("Red", "Orange", "Yellow", "Blue", "Green"),
  Price = c(1.20, 1.40, 1.70, 3.50, 4.40),
  In_Stock = c("Yes", "Yes", "No", "Yes", "No")
)

table_b <- tibble(
  SKU = c(102345, 105432, 106789, 104567, 107654),
  Fruit = c("Apple", "Banana", "Grape", "Orange", "Pear"),
  Color = c("Red", "Yellow", "Purple", "Orange", "Green"),
  Sale_Price = c(1.00, 0.50, 2.00, 1.20, 1.10),
  Number_in_Stock = c(50, 120, 0, 75, 0)
)
head(table_a)
head(table_b)
```

### a) Left join

```r
table_ab <- table_a %>%
left_join(table_b, by = c("SKU", "Fruit", "Color")) %>%
select("SKU", "Fruit", "Color", "Price", "Sale_Price", "In_Stock", "Number_in_Stock")
head(table_ab)
```

### b) Right Join

```r
table_ab <- table_a %>%
right_join(table_b, by = c("SKU", "Fruit", "Color")) %>%
select("SKU", "Fruit", "Color", "Price", "Sale_Price", "In_Stock", "Number_in_Stock")
head(table_ab)
```

### c) Inner Join

```r
table_ab <- table_a %>%
inner_join(table_b, by = c("SKU", "Fruit", "Color")) %>%
select("SKU", "Fruit", "Color", "Price", "Sale_Price", "In_Stock", "Number_in_Stock")
head(table_ab)
```

### d) Full Join

```r
table_ab <- table_a %>%
full_join(table_b, by = c("SKU", "Fruit", "Color")) %>%
select("SKU", "Fruit", "Color", "Price", "Sale_Price", "In_Stock", "Number_in_Stock")
head(table_ab)
```

### e) Semi Join

```r
table_ab <- table_a %>%
semi_join(table_b, by = c("SKU", "Fruit", "Color"))
head(table_ab)
```

```r
table_ab <- table_b %>%
semi_join(table_a, by = c("SKU", "Fruit", "Color"))
head(table_ab)
```

### f) Anti Join

```r
table_ab <- table_a %>%
anti_join(table_b, by = c("SKU", "Fruit", "Color"))
head(table_ab)
```

```r
table_ab <- table_b %>%
anti_join(table_a, by = c("SKU", "Fruit", "Color"))
head(table_ab)
```
